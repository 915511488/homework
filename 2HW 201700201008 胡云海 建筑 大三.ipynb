```python
t = input("请输入一段字符串：")                         # 输入
t = t.lower()                                         # 不区分大小写
d = {}
for i in t:
    if i in "abcdefghijklmnopqrstuvwxyz":
        d[i] = d.get(i,0) + 1                         # 用字典计数对应的字母
ls = list(d.items()) 
ls.sort(key = lambda x:x[1],reverse = True)           # 频率大小排序
for j in range(len(d)):
    word,count = ls[j]                                 
    print("{:<10}{:<5}".format(word,count))            # 输出
```

```python
def circu(s):                                                      # 定义函数
    d={}
    for i in ('数字','空格','字母','其他字符'):                      # 初始化字典
        d[i] = d.get(i,0)       
    for i in s:                                                    # 识别计算
        if i.isdigit():
            d['数字'] = d.get('数字')+1
        elif i.isspace():
            d['空格'] = d.get('空格')+1
        elif i.isalpha():
            d['字母'] = d.get('字母')+1
        else:
            d['其他字符'] = d.get('其他字符')+1
    ls = list(d.items())
    ls.sort(key = lambda x:x[1],reverse = True)                   # 频率大小排序
    for i in range(len(ls)):
        typer,count=ls[i]
        print("{0:{2}<15}{1:<5}".format(typer,count,chr(12288)))    # 格式化输出
    return ls
    
end = circu(input("请输入一段字符串:"))                             # 主函数
```

```python
def move(n, Afrom, Bto) :
  global i
  print ("第%d步:将%d号盘子从%s ——> %s" %(i, n, Afrom, Bto))
  i += 1

def hanoi(n, A, B, C) :
  if n == 1 :
    move(1, A, C)
  else :
    hanoi(n - 1, A, C, B) 
    move(n, A, C)    
    hanoi(n - 1, B, A, C)

i=1
n = int(input('请输入盘子数'))
print ("汉尼拔移动步骤如下：")
hanoi(n, 'A', 'B', 'C')

```
